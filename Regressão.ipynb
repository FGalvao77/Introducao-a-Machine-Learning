```
# importando as bibliotecas
import numpy as np                                # gerar matrizes
from sklearn.linear_model import LinearRegression # algoritmo para realizar nossas predições
import matplotlib.pyplot as plt                   # gerar gráficos

# exibir gráficos
%matplotlib inline                                
```

```
# instanciando os dados da variável independente - x (explicativa)
x = np.array([10, 11, 12, 13, 14, 15]).reshape(- 1, 1) # horas de estudo
```

```
# exibindo as dados da variável
x
```

**Obs:** A função `reshape(-1, 1)` ela tem a função de remodelar os dados para ter u**ma coluna e diversas linhas**.

O predcitor(x) necessita **ser uma matriz de matrizes** e a resposta (y) **pode ser uma matriz simples.**

```
# instanciando os dados da variável dependente - y (variável a ser predita)
y = np.array([53, 52, 60, 63, 72, 70]) # pontos do exame (0 -100)
```

```
# exibindo as dados do y
y
```

```
# definindo o modelo preditivo com o algoritmo "LinearRegression"
modelo = LinearRegression()
```

```
# treinando o modelo com os dados instanciados
modelo.fit(x, y)
modelo
```

```
# realizando as predições e instaciando na variável "y_pred"
y_pred = modelo.predict(x)
y_pred
```

```
# configurando o gráfico e plotando o gráfico
plt.scatter(x, y)
plt.plot(x, y_pred)
plt.xlabel('Horas de estudo')
plt.ylabel('Pontos no exame')
plt.show()
```

`Geramos o gráfico usando a função`**.scatter()**, `que recebe o` **x** e o **y** `como argumentos.`

```
# obtendo o coeficientes da variáveies "x" e "y"
print(modelo.coef_)       # coeficiente de inclinação da reta
print(modelo.intercept_)  # ponto de interceptação da reta no eixo y
```

### Vamos fazer um exercício prático para um melhor entendimento da aplicação da equação.

Dado um caso, onde um aluno estudou cerca de `16 horas`, qual é a sua estimativa de pontos no exame?

Iremos realizar a predição na mão.

    y = 4,23x + 8,80
    y = 4,23(16) + 8,80
    y = 76,48

**Com essas horas de estudo, o aluno provavelmente terá em torno de 76 pontos no exame.**

```
# agora vamos usar o modelo gerado para realizar predição
# atenção! é necessário entregar os dados no formato de matriz (array)
x_new = np.array([16]).reshape(-1, 1) # realizando o processamento do dado
x_new
```

```
# realizando a predição
y_new = modelo.predict(x_new)
print(f'A nota estimada é de ===> {y_new}')
```

## Como avaliar se o nosso modelo está realizando as predições?

Para isso, utilizaremos o `R2 (coeficiente de determinação)` através do método **.score()**.

Onde quanto mais pŕoximo de 1,0  indica a **eficiência** do modelo. Lembrando que, também podemos ter valores negativos.

```
# aplicando a função ".score()" com os argumentos de "x" e "y"
modelo.score(x, y)
```

Portando nosso modelo possui cerca de **89%** de acurácia.
